```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/arsi/Documents/Coding/machineLearning/Employee.csv')
```

```python
df['Gender'] = df['Gender'].replace({'Male': 1, 'Female': 0})
df['EverBenched'] = df['EverBenched'].replace({'Yes': 1, 'No': 0})
df_dumm = pd.get_dummies(df['Education'], drop_first=True).astype(int)
df = pd.concat([df, df_dumm], axis=1).drop(columns=['Education'])
df.head()
```

```python
X = df[['Age','Gender','ExperienceInCurrentDomain', 'PaymentTier', 'EverBenched', 'JoiningYear', 'Masters', 'PHD']]
y = df['LeaveOrNot']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 22)
```

```python
X_train_scaled = X_train.copy()
X_train_scaled['Age'] = X_train['Age']/100
X_train_scaled['JoiningYear'] = X_train['JoiningYear'] - 2011
X_train_scaled['JoiningYear'] = X_train_scaled['JoiningYear']/10
X_train_scaled['PaymentTier'] = X_train['PaymentTier']/10
X_train_scaled['ExperienceInCurrentDomain'] = X_train['ExperienceInCurrentDomain']/10


X_test_scaled = X_test.copy()
X_test_scaled['Age'] = X_test['Age']/100
X_test_scaled['JoiningYear'] = X_test['JoiningYear'] - 2011
X_test_scaled['JoiningYear'] = X_test_scaled['JoiningYear']/10
X_test_scaled['PaymentTier'] = X_test['PaymentTier']/10
X_test_scaled['ExperienceInCurrentDomain'] = X_test['ExperienceInCurrentDomain']/10

X_train_scaled.head()
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100, input_shape=(8,), activation='relu', kernel_initializer='ones', bias_initializer='zeros'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])


model.compile(optimizer = 'Adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
model.fit(X_train_scaled, y_train, epochs= 1000)
```

```python
model.evaluate(X_test_scaled, y_test)
```
